1. What is the difference between a function and a method in Python?

-> Function:
a. They are	defined using def outside a class
b. They are called directly using function_name()
c. They are	not bound to any object
d. They are general-purpose used independently
Example:
def add(a, b):
    return a + b

Method:
a. They are defined inside a class and operates on an object
b. They are called on an object using object.method_name()
c. They are bound to an instance of a class (self)
d. They are	used in OOP and operates on class instances
Example:

    def add(self, a, b):
        return a + b
obj = Math()   
print(obj.add(2, 3))

3. What are the different ways to define and call a function in Python?

-> a. Regular Function Definition & Call:

The most common way to define a function using def and call it by its name.

Definition:
def greet(name):
    return f"Hello, {name}!"
Calling:
print(greet("Alice"))  # Output: Hello, Alice!

b. Function with Default Parameters:

You can set default values for parameters.

Definition:

    return f"Hello, {name}!"
Calling:

c. Function with *args (Variable-Length Positional Arguments):

Allows passing multiple arguments as a tuple.

Definition:

    return sum(args)
Calling:

print(addNumbers(10, 20))      # Output: 30

d. Function with **kwargs (Variable-Length Keyword Arguments):

Accepts multiple keyword arguments as a dictionary.

Definition:

    for key, value in kwargs.items():
        print(f"{key}: {value}")
Calling:
userInfo(name="Alice", age=25, city="New York")
""" Output:
name: Alice
age: 25
city: New York"""

e. Lambda (Anonymous) Functions:

Lambda functions are single-expression functions.

Definition:

Calling:
print(add(5, 3))  # Output: 8

f. Function as an Argument (Higher-Order Functions):

Functions can be passed as arguments to other functions.

Definition:

    return func(x, y)
def multiply(a, b):
    return a * b
Calling:
result = applyOperation(3, 4, multiply)
print(result)  # Output: 12

g. Recursive Function:

A function calling itself.

Definition:
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
Calling:
print(factorial(5))  # Output: 120

h. Nested Functions:

A function inside another function.

Definition:
def outerFunction(text):
    def innerFunction():
        return f"Inner says: {text}"

Calling:

5. What are iterators in Python and how do they differ from iterables?

-> An iterable is any Python object that can be looped over using a for loop. Examples include:
a. Lists, tuples, sets, dictionaries, strings
b. Objects that implement the __iter__() method

Example:
myList = [1, 2, 3, 4]
for item in myList:  # my_list is iterable
    print(item)

An iterator is an object that remembers its position and provides elements one at a time when asked.
It implements two special methods:
a. __iter__():Returns the iterator object itself.
b. __next__():Returns the next item in the sequence and raises StopIteration when exhausted.

Example:
myList = [1, 2, 3, 4]
iterator = iter(myList)  # Convert iterable to iterator
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
print(next(iterator))  # Output: 3
print(next(iterator))  # Output: 4
print(next(iterator))  # Raises StopIteration

6. Explain the concept of generators in Python and how they are defined.

-> A generator in Python is a special type of function that produces values lazily (one at a time) instead of returning them all at once. This makes generators memory efficient and fast for handling large data streams.

Generators are defined using the yield keyword inside a function instead of return.
Example:
def simpleGenerator():
    yield 1
    yield 2
    yield 3
gen = simpleGenerator() # Using the generator
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
print(next(gen))  # Raises StopIteration

7. What are the advantages of using generators over regular functions?

-> The advantages of using generators over regular functions :

Memory Efficient - Uses yield to generate values lazily, avoiding large memory consumption.
Example:
def squaresList(n): # Regular function (stores all squares in memory)
    return [x**2 for x in range(n)]
print(squaresList(1000000))  # High memory usage

def squaresGenerator(n): # Generator function (produces squares one by one)
    for x in range(n):
        yield x**2
gen = squaresGenerator(1000000)  # No large memory allocation
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1

Faster Execution - Avoids unnecessary computations, runs only when needed.
Example:
def readFile(filename):
    with open(filename, 'r') as file:
        return file.readlines()  # Loads all lines at once

def readFileGen(filename): # Generator function (reads one line at a time)
    with open(filename, 'r') as file:
        for line in file:
            yield line  # Returns one line at a time
for line in readFileGen("large_file.txt"): # Using generator
    print(line.strip())  # Efficient processing

Automatic State Management - Remembers the last position, no need for external tracking.
Example:
def countdown(n):
    while n > 0:
        yield n
        n -= 1  # State is automatically saved
gen = countdown(5)
print(next(gen))  # Output: 5
print(next(gen))  # Output: 4
print(next(gen))  # Output: 3

8. What is a lambda function in Python and when is it typically used?

-> A lambda function in Python is a small, anonymous function that can have any number of arguments but only one expression.

It is often used for short, simple operations where defining a full function with def would be unnecessary. They are also used as arguments in map(), filter(), and reduce() as well as inside list comprehensions and inside sorting functions (sorted()).

Example:

9. Explain the purpose and usage of the `map()` function in Python.

-> The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results. 
It is used to transform each item in an iterable by applying the specified function.

Example:
nums = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, nums)
print(list(squared))  # Output: [1, 4, 9, 16]

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

-> map(): Applies a function to all elements in an iterable.
filter(): Selects elements that satisfy a condition.
reduce(): Performs a cumulative operation on a list.

Example:
nums = [1, 2, 3, 4]

squared = list(map(lambda x: x ** 2, nums)) # map()
print(squared)  # Output: [1, 4, 9, 16]

evens = list(filter(lambda x: x % 2 == 0, nums)) # filter()
print(evens)  # Output: [2, 4]

from functools import reduce
sumAll = reduce(lambda x, y: x + y, nums) # reduce()
print(sumAll)  # Output: 10

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sumEven(listNumbers):
    sum=0
    for i in listNumbers:
        if i%2==0:
            sum=sum+i
    return sum
numbers=list()
length=int(input("Enter the length of the list:"))
for i in range(0,length):
    number=int(input(f"Enter number {i+1} : "))
    numbers.append(number)
print(f"List of numbers is {numbers}")
print(f"Sum of all even numbers in the list is {sumEven(numbers)}")
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.

def reverseString(string):
    revString=""
    i=len(string)-1
    while i>=0:
        revString=revString+string[i]
        i=i-1
    return revString
string=input("Enter a string value:")
print(f"Reverse of {string} is {reverseString(string)}")
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def squareNumbers(listNumbers):
    squareList=[]
    for i in listNumbers:
        squareList.append(i**2)
    return squareList
numbers=list()
length=int(input("Enter the length of the list:"))
for i in range(0,length):
    number=int(input(f"Enter number {i+1} : "))
    numbers.append(number)
print(f"List of numbers is {numbers}")
print(f"List containing the squares of each number is {squareNumbers(numbers)}")
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def checkPrime(number):
    for i in range(2,(number//2)+1):
        if number%i==0:
            return False
    return True
number=int(input("Enter a number between 1 and 200:"))
if number>200 or number<1:
    print("The entered number should be between 1 and 200.")
else:
    if checkPrime(number)==True:
        print(f"{number} is a prime number")
    else:
        print(f"{number} is not a prime number")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self,maxTerm):
        self.maxTerm=maxTerm
        self.firstNum=0
        self.secondNum=1
        self.count=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count>=self.maxTerm:
            raise StopIteration
        fibonacciNumber=self.firstNum
        self.firstNum,self.secondNum=self.secondNum,self.firstNum+self.secondNum
        self.count=self.count+1
        return fibonacciNumber
term=int(input("Enter the maximum term to generate the Fibonacci sequence:"))
print(f"Fibonacci sequence upto {term} terms are :")
for i in Fibonacci(term):
    print(i)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def generatePowers(exponent):
    for i in range(exponent+1):
        yield 2**i
exponent=int(input("Enter an exponent to yield the powers of 2:"))
print(f"Powers of 2 upto given exponent {exponent} are :")
for i in generatePowers(exponent):
    print(i)
```
